```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
```

```python
x = np.array([1,1.5,3,5,3.5,4.5,3.5])
y = np.array([1,2,4,7,5,5,4.5])
```

```python
data = np.column_stack((x,y))
print(data)
l = len(data)
print(l)
```

```python
l = len(data)
inertias = []
for i in range(1,l):
  kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)
  kmeans.fit(data)
  inertias.append(kmeans.inertia_)

plt.plot(range(1,l), inertias, marker='o',linestyle='-')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2, random_state=42,n_init=10)
kmeans.fit(data)
plt.scatter(x, y,c=kmeans.labels_, cmap='viridis',alpha=0.7,label="Clustered Points")
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:,0], centroids[:,1], c='red', marker ='X', s=200, label="Centroids")
plt.title('K-Means Clustering')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.patches as mpatches

data = pd.read_csv('WisconsinCancer.csv')
data.head()
```

```python
data.info()
```

```python
data = data.drop(['id','Unnamed: 32'], axis=1)
data.isnull().sum()
```

```python
data['diagnosis'] = data['diagnosis'].map({'M':1, 'B': 0})
features = ['radius_mean','texture_mean']
data_selected = data[features]
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_selected)
```

```python
wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init='k-means++',random_state=42)
  kmeans.fit(data_scaled)
  wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(range(1,11),wcss, marker='o',linestyle='--')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method for Optimal Clusters')
```

```python
kmeans = KMeans(n_clusters=2, init='k-means++', random_state=42)
data['Cluster'] = kmeans.fit_predict(data_scaled)
centroids = kmeans.cluster_centers_

cluster_labels = {data['Cluster'][data['diagnosis']==0].mode()[0]:'Benign',
                  data['Cluster'][data['diagnosis'] == 1].mode()[0]: 'Malignant'}
data['Cluster_Label'] = data['Cluster'].map(cluster_labels)
plt.figure(figsize=(12,5))
plt.subplot(1,2,2)
colors = ['yellow' if label == 'Benign'else 'purple' for label in data['Cluster_Label']]
plt.scatter(data_scaled[:,0], data_scaled[:,1], c=colors, alpha=0.6, edgecolors='k')
plt.scatter(centroids[:,0],centroids[:,1], c='red', marker = 'X', s=200, label = 'Centroids')
plt.xlabel('Standarized Radius Mean')
plt.ylabel('Standarized Texture Mean')
plt.title('K-Means Clustering on Radius Mean & Texture Mean')

benign_patch = mpatches.Patch(color='yellow',label='Benign')
malignant_patch = mpatches.Patch(color='purple', label = 'Malignant')
plt.legend(handles = [benign_patch, malignant_patch, plt.Line2D([0],[0], marker = 'X', color = 'red', markersize =10, label='Centroids')])

plt.show()
```
