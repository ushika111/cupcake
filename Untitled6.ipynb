```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import statsmodels.api as sm
df = pd.read_csv('HDI.csv')
df.head()
```

```python
x = df['Human Development Index']
y = df['Mean Mathematics Performance Score']
plt.scatter(x,y, color ='blue', alpha = 0.5)
plt.xlabel('Human Development Index')
plt.ylabel('Mean Mathematics Performance Score')
plt.title('Scatter Plot of HDI vs Mean Performance')
plt.show()
```

```python
result = stats.linregress(x, y)
print("LinregressResult:")
print(f"slope: {result.slope}")
print(f"Intercept: {result.intercept}")
print(f"R-value: {result.rvalue}")
print(f"P-value: {result.pvalue}")
print(f"STandard Error: {result.stderr}")
```

```python
x_with_constant = sm.add_constant(x)
model = sm.OLS(y, x_with_constant).fit()
print(model.summary())
```

```python
residuals = model.resid
fitted_values = model.fittedvalues
plt.figure(figsize=(6,4))
plt.scatter(x, residuals, alpha=0.5)
plt.axhline(y=0, color = 'red', linestyle='--')
plt.xlabel('Human Development Index Mean')
plt.ylabel('Residuals')
plt.title('Residuals vs HDI Mean')
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(fitted_values, residuals, alpha=0.5)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Predicted Mathematics Performance Score')
plt.ylabel('Residuals')
plt.title('Residuals vs Predicted Values')
plt.show()
```

```python
sm.qqplot(residuals, line='45', fit = True)
plt.title('Normal Probability Plot of Residuals')
plt.show()
```
